```python

#  Step 1: Install Required Package


# !pip install ibm-watsonx-ai

#  Step 2: Import Libraries

from ibm_watsonx_ai.foundation_models import Model
from ibm_watsonx_ai import Credentials, WatsonxAI
import json



creds = Credentials(
    url="https://us-south.ml.cloud.ibm.com",  # You may change region if needed
    api_key="your_ibm_api_key_here"
)

watsonx = WatsonxAI(credentials=creds)


#  Step 4: Load Granite Model

model = Model(
    model_id="ibm/granite-13b-instruct-v2",  # Use updated if instructed
    params={"decoding_method": "greedy", "max_new_tokens": 300},
    watsonx_ai=watsonx
)

# Step 5: Create Prompt Template

def generate_prompt(symptom_input: str):
    return f"""
You are a health assistant. A user said: "{symptom_input}"

Based on WHO and medical guidelines, provide:
1. Probable Conditions
2. Urgency Level (Mild / Moderate / Severe)
3. Home Remedies
4. When to Consult a Doctor

Return output in this JSON format:
{{
  "conditions": [...],
  "urgency": "...",
  "remedies": [...],
  "doctor_advice": "..."
}}
"""

#  Step 6: Run Prediction

def analyze_symptoms(symptom_input: str):
    prompt = generate_prompt(symptom_input)
    result = model.generate(prompt)
    return result['generated_text']


#  Step 7: Test with Example

# You can replace this with Hindi or any other language input too.
symptom = "I have a sore throat and fever"
output = analyze_symptoms(symptom)

print("üîé AI Output:")
print(output)

# Optional: Parse JSON if format is correct
try:
    response = json.loads(output)
    print("\n‚úÖ Parsed JSON Output:")
    print(json.dumps(response, indent=2))
except json.JSONDecodeError:
    print("\n‚ö†Ô∏è Output is not valid JSON. You may adjust the prompt or parse manually.")
```
